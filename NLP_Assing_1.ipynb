###### The final dataset before Text preprocessing

```python
import re
import emoji
from textblob import TextBlob
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer,WordNetLemmatizer
```

###### 1.Here we are using `regex` for removing unnecessary characters in text 
###### 2. `emoji` using this we are decoding the information from emoji present in the text
###### 3. `TextBlob` using textblob we are correcting the spelling of english words
###### 4. `Stem` using stemming we are storing the root word of a particular word 
###### 5. `tokenize` using this dividing paragraph or sentences  into tokens

```python
stp=stopwords.words('english')
```

```python
'not' in stp
```

```python
stp.remove('not')
```

###### removing `not` in the stopwords 

```python
new0=[]
for sent in data['text']:
    new1=[]
    new=sent.lower()              #first converting into lower case
    new=emoji.demojize(new)       #getting desired information from the emojis
    new=TextBlob(new).correct().string    #correcting the spelling mistakes 
    new=re.sub('<.*?>','',new)            # removing unnecessary characters
    new=re.sub('https?://\S*|www\S*','',new)
    new=re.sub("[^a-z ]","",new)
    for word in word_tokenize(new):      #using word_tokenize dividing into words
        if word in stp:                  #neglecting the words which are in the stopwords 
            pass
        else:
            stem=WordNetLemmatizer().lemmatize(word) #stemming of each word which are unique
            new1.append(stem)
    new0.append(' '.join(new1))                     #Appending into the list
```

```python
cdata=pd.DataFrame({'cleaned_text':new0})
```

```python
cdata
```
